```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB

df = pd.read_excel(r"C:\Users\md747\Downloads\1750577562424-iris-train.xlsx")
df
```

```python
x = df.iloc[:,:-1]
x
```

```python
y = df.iloc[:,-1]
```

```python
# y
```

```python
model = GaussianNB()
# x.values
model.fit(x.values,y.values)
```

```python
predictions = model.predict([[3.4,	1.0,	4.5,	0.5]])
```

```python
print(predictions)
```

```python
# lets predict on my test dataset
test_dataset = pd.read_excel(r"C:\Users\md747\Downloads\1750577561477-iris-test.xlsx")
```

```python
x_test= test_dataset.iloc[:,:-1]
```

```python
predictions = model.predict(x_test.values)
```

```python
# predictions
```

```python
# checking accuracy of my model for test set
y_test = test_dataset.iloc[:,-1]
```
