```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree #for making graph
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
features = [
    [150, 0], [170, 0], [160, 0],
    [130, 1], [120, 1], [140, 1]
]

labels = ["apple", "apple", "apple", "orange","orange","orange"]

clf = tree.DecisionTreeClassifier(random_state=0)

clf = clf.fit(features,labels)


ans = clf.predict([[160,0]])
print(f"output is {ans}")


```

```python
plt.figure(figsize=(15,6))
plot_tree(clf, feature_names=["weight", "color"], class_names=["apple", "orange"], filled=True)
plt.show()
```
